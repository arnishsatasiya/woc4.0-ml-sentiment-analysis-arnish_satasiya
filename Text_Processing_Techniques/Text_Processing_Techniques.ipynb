```python
import pandas as pd
import numpy as np
import nltk
nltk.download()
```

```python
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/data.csv')
data
```

# Tokenization

```python
from nltk.tokenize import  word_tokenize
```

```python
def tokenize_text(x):
  x=word_tokenize(x)
  return x

data['Review']=data['Review'].apply(tokenize_text)
data
```

# Removing StopWords

```python
from nltk.corpus import stopwords
nltk_stopwords = stopwords.words('english')
nltk_stopwords
```

```python
def remove_stopwords(x):
  x=[word for word in x if word not in nltk_stopwords]
  return x

data['Review']=data['Review'].apply(remove_stopwords)
data
```

# Stem all the words using PorterStemmer.


```python
from nltk.stem import PorterStemmer
```

```python
def stem_words(x):
  porter=PorterStemmer()
  list=[]
  for word in x:
     list.append(porter.stem(word))
  return ' '.join(list)

data['Review']=data['Review'].apply(stem_words)
data
```
