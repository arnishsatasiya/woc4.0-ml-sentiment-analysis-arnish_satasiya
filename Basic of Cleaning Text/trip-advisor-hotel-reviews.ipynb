```python
data = pd.read_csv('../input/trip-advisor-hotel-reviews/tripadvisor_hotel_reviews.csv')
data
```

```python
data.isna().sum()
```

```python
def update_rating(x):
    if x<=2:
        return 0
    elif x==3:
        return 1
    else:
        return 2
 

data['Rating']=data['Rating'].apply(update_rating)
data
    
```

```python
def modify(x):
    x=x.lower()
    x=re.sub(r"https?://[a-z0-9./]+", ' ', x)
    x=re.sub(r'[^a-z]',' ',x)
    x=re.sub(r' +',' ',x)
    return x

data['Review']=data['Review'].apply(modify)
data
```
