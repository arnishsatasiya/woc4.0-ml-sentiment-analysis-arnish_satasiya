```python
def update_rating(x):
    if x<=2:
        return 0
    elif x==3:
        return 1
    else:
        return 2
 

data['Rating']=data['Rating'].apply(update_rating)
data
    
```

**Clean the text to achieve the following:**


1.   Lower all the text.
2.   Substitute all the non alphabetical characters(a-z) by â€˜ â€˜ (space).
3. Remove all the hyperlinks if present in the text.
4. If there are multiple consequent spaces in the text, replace them by a single space (for example â€˜    â€˜ to â€˜ â€˜ ). 






```python
def modify(x):
    x=x.lower()
    x=re.sub(r"https?://[a-z0-9./]+", ' ', x)
    x=re.sub(r'[^a-z]',' ',x)
    x=re.sub(r' +',' ',x)
    return x

data['Review']=data['Review'].apply(modify)
data
```

```python
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
```

# Tokenization

```python
from nltk.tokenize import  word_tokenize
```

```python
def tokenize_text(x):
  x=word_tokenize(x)
  return x

data['Review']=data['Review'].apply(tokenize_text)
data
```

# Removing StopWords

```python
from nltk.corpus import stopwords
nltk_stopwords = stopwords.words('english')
nltk_stopwords
```

```python
def remove_stopwords(x):
  x=[word for word in x if word not in nltk_stopwords]
  return x

data['Review']=data['Review'].apply(remove_stopwords)
data
```

# Stem all the words using PorterStemmer.

```python
from nltk.stem import PorterStemmer
```

```python
def stem_words(x):
  porter=PorterStemmer()
  list=[]
  for word in x:
     list.append(porter.stem(word))
  return ' '.join(list)

data['Review']=data['Review'].apply(stem_words)
data
```
